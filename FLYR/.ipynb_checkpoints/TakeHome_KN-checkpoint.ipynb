### EDA 

```python
# First, convert fare to USD 
top_10_currency=search_df.groupby(search_df.currency).search_id.count().nlargest(10, 'first')
```

```python
top_10_currency.index
```

```python
#convert currency into USD, this is now our target 
#ep = EasyPeasy(fuzzy_threshold=True)
converted_fare_USD=[] 
for i in range(len(search_df)):
    if search_df.currency[i] in top_10_currency.index: 
        if search_df.currency[i]=='USD':
            converted_fare_USD.append(search_df.fare[i])
        elif search_df.currency[i]=='CAD':
            converted_fare_USD.append(search_df.fare[i]*0.82)
        elif search_df.currency[i]=='EUR':
            converted_fare_USD.append(search_df.fare[i]*1.2)
        elif search_df.currency[i]=='ISK':
            converted_fare_USD.append(search_df.fare[i]*0.0093)
        elif search_df.currency[i]=='GBP':
            converted_fare_USD.append(search_df.fare[i]*1.35)
        elif search_df.currency[i]=='SEK':
            converted_fare_USD.append(search_df.fare[i]*0.13)
        elif search_df.currency[i]=='NOK':
            converted_fare_USD.append(search_df.fare[i]*0.13)
        elif search_df.currency[i]=='DKK':
            converted_fare_USD.append(search_df.fare[i]*0.16)
        elif search_df.currency[i]=='AUD':
            converted_fare_USD.append(search_df.fare[i]*0.79)    
        elif search_df.currency[i]=='JPY':
            converted_fare_USD.append(search_df.fare[i]*0.0089)   
        else:
            converted_fare_USD.append(search_df.fare[i])
            
    else:
        converted_fare_USD.append(search_df.fare[i]*0.27)
 
# some bug in the library 
# search_df['converted_fare_USD'] = search_df.apply(lambda x: ep.currency_converter(amount=x.fare,\
#                                                                                   from_currency=x.currency,\
#                                                                                   to_currency="USD",pretty_print=True ), axis=1 )
```

```python
#time of the day departure 
hour= search_df.departure_datetime_1.apply(lambda x: int(x.split(';')[0].split('T')[1].split(':')[0]))
```

```python
search_df.arrival_datetime_1[250]
```

```python
time_of_day=[] 
for i in range(len(hour)):
    if hour[i]<7:#early morning  
        time_of_day.append(1)  
    elif hour[i]>=7 and hour[i]< 12: #morning
        time_of_day.append(2)
    elif hour[i]>=12and hour[i]< 17: #afternoon 
        time_of_day.append(3)
    elif hour[i]>=17 and hour[i]< 20: #evening 
        time_of_day.append(4)
    else:  #late evening 
        time_of_day.append(5) 
```

```python
search_df['time_of_day']=time_of_day 
```

```python
search_df.converted_fare_USD.groupby(search_df.time_of_day).mean().plot.bar()
plt.title('fare by departure time of day')
#late evening and afternoon are costly  
```

```python
#season 
month= search_df.departure_datetime_1.apply(lambda x: int(x.split(';')[0].split('T')[0].split('-')[1]))
```

```python
season=[] 
for i in range(len(month)): 
    if month[i] in (3,4, 5):  #spring 
        season.append(1)  
    elif month[i] in (6,7, 8): #summer  
        season.append(2)
    elif month[i] in (9,10, 11):  
        season.append(3)
    else:  
        season.append(4) 
```

```python
search_df['season']=season 
```

```python
search_df.converted_fare_USD.groupby(search_df.season).mean().plot.bar()
plt.title('fare by season')
#summer/winter (holiday) is more expensive!
```

```python
#last arrival time 
last_arrival=[] 
for i in range(len(search_df)):
    if type(search_df.arrival_datetime_1[i])!=float:
        last_arrival.append(search_df.arrival_datetime_1[i].split(';')[-1])
    else: 
        last_arrival.append(np.NaN)
```

```python
first_departure= search_df.departure_datetime_1.apply(lambda x: x.split(';')[0])
```

```python
len(first_departure), len(last_arrival)
```

```python
time_1=[] 
for i in range(len(search_df)):
    if search_df.arrival_datetime_1[i] <'2560' and type(search_df.arrival_datetime_1[i])!=float:
        time_1.append(pd.Timedelta(pd.to_datetime(last_arrival[i])-pd.to_datetime(first_departure[i])).seconds//3600)
    else:
        time_1.append(np.NaN)    
```

```python
search_df['trip_duration_hours']=time_1
```

```python
#plt.plot(search_df.trip_duration_hours, search_df.converted_fare_USD)
```

```python
#create miles bins  
miles_bins=[]
for i in range(len(search_df)):
    if search_df.miles[i]<10: 
        miles_bins.append(1)
    elif search_df.miles[i]<20: 
        miles_bins.append(2)
    elif search_df.miles[i]<30: 
        miles_bins.append(3)
    elif search_df.miles[i]<40: 
        miles_bins.append(4)
    else:
        miles_bins.append(5) 
```

```python
#create hour bins  
hours_bins=[]
for i in range(len(search_df)):
    if search_df.trip_duration_hours[i]<2: 
        hours_bins.append(1)
    elif search_df.trip_duration_hours[i]<4: 
        hours_bins.append(2)
    elif search_df.trip_duration_hours[i]<6: 
        hours_bins.append(3)
    elif search_df.trip_duration_hours[i]<8: 
        hours_bins.append(4)
    else:
        hours_bins.append(5) 
```

```python
search_df['miles_bins']=miles_bins 
```

```python
search_df['hours_bins']=hours_bins 
```

```python
fig = plt.figure(figsize=(5,5))
search_df.converted_fare_USD.groupby([search_df.miles_bins, search_df.hours_bins]).mean().plot.bar()
plt.title('fare by miles and trip duration')
```

### Features

```python
#convert object fields into numbers 
#object_cols=search_df[search_df.dtypes=='object'] 
search_df['partner_num']=pd.Categorical(search_df.partner_id).codes
search_df['pos_num']=pd.Categorical(search_df.pos).codes
search_df['cabin_num']=pd.Categorical(search_df.cabin_class.str.lower()).codes
search_df['airline_num']=pd.Categorical(search_df.dominate_airline).codes
search_df['roundtrip']=pd.Categorical(search_df.roundtrip).codes
search_df['miles_per_hour']=search_df.miles/search_df.trip_duration_hours
```

```python
miles_per_hour=[]
for i in range(len(search_df)):
    if search_df.trip_duration_hours[i]>0:
        miles_per_hour.append(search_df.miles[i]/search_df.trip_duration_hours[i])
    else: 
        miles_per_hour.append(0)
    
```

```python
search_df['miles_per_hour']=miles_per_hour 
```

### Modeling

```python
y=df_features.converted_fare_USD
#X=df.drop(['diagnosis'], axis = 1, inplace = False, errors = 'ignore')
X=df_features[['partner_num', 'pos_num', \
          'num_layover','cabin_num', 'season', 'time_of_day',\
               'miles_per_hour', 'miles', 'trip_duration_hours','roundtrip', 'passengers', 'num_requests','airline_num']]
X_train,X_test, y_train, y_test= train_test_split(X,y)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#normalize/standarize features
min_max_scaler = MinMaxScaler()
X_train_norm= min_max_scaler.fit_transform(X_train)
X_test_norm= min_max_scaler.fit_transform(X_test)
```

```python
## Logistic Regression 
lr=LinearRegression() 
lr.fit(X_train_norm, y_train)
y_pred=lr.predict(X_test)
lr.score(X_test_norm, y_test)
```

```python
accuracy = r2_score(y_test, y_pred)
print 'Cross-Predicted Accuracy:', accuracy
```

```python
lr.coef_
```

```python
## Random Forest 
rf=RandomForestRegressor()
rf.fit(X_train_norm, y_train)
y_pred=rf.predict(X_test_norm) 
rf.score(X_test_norm, y_test)
```

```python
accuracy = r2_score(y_test, y_pred)
print 'Cross-Predicted Accuracy:', accuracy
```

```python
#f= np.array(zip(X.columns, clf.feature_importances_))
#f[::, f[0,].argsort()[::-1]] 
important_features = pd.Series(data=rf.feature_importances_,index=X.columns)
important_features.sort_values(ascending=False,inplace=True)
important_features
```

```python
tree=DecisionTreeRegressor()
tree.fit(X_train_norm, y_train)
y_pred=tree.predict(X_test_norm) 
tree.score(X_test_norm, y_test)
```

### Modeling 

```python
y=df['booked']
#X=df.drop(['diagnosis'], axis = 1, inplace = False, errors = 'ignore')
X=df_features[['partner_num', 'pos_num', 'cabin_num', 'passengers', 'num_requests','airline_num', 'converted_fare_USD']]
X_train,X_test, y_train, y_test= train_test_split(X,y)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
lr.score(X_test, y_test)
```

```python
#since it's unbalanced dataset, we would need to look at confusion matrix, f1 score 
acurracy= accuracy_score(y_test, y_pred)
precision= precision_score(y_test, y_pred)
recall= recall_score(y_test, y_pred)
f1= f1_score(y_test, y_pred)

print "Logistic Regression Accuracy:{0:0.3f}, Precision:{0:0.3f}, Recall:{0:0.3f}, F1:{0:0.3f}".format(acurracy, precision,recall, f1 )
```
