```python
#function to generate bootstrap samples of the data
def bootstrap_func(df, n=100, f=0.8):
    """
    function to generate bootstrap samples of the data
    input: df, n=number of samples, f= fraction 
    output: df samples
    """
    samples={}
    for i in range(n):
        samples[i]=df.sample(frac=f, replace=False) 
    return samples
```

```python
y=df.pop('diagnosis_num')
#X=df.drop(['diagnosis'], axis = 1, inplace = False, errors = 'ignore')
X=df_features.drop(['diagnosis_num'], axis = 1, inplace = False, errors = 'ignore') #use the subset of features(feature reduction)
X_train,X_test, y_train, y_test= train_test_split(X,y)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
## Logistic Regression 
lr=LogisticRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
lr.score(X_test, y_test)
```

```python
#since it's unbalanced dataset, we would need to look at confusion matrix, f1 score 
acurracy= accuracy_score(y_test, y_pred)
precision= precision_score(y_test, y_pred)
recall= recall_score(y_test, y_pred)
f1= f1_score(y_test, y_pred)

print "Logistic Regression Accuracy:{0:0.3f}, Precision:{0:0.3f}, Recall:{0:0.3f}, F1:{0:0.3f}".format(acurracy, precision,recall, f1 )
```

```python
loo = LeaveOneOut()
X_array=X.values
loo.get_n_splits(X_array)
loo_scores=[]
for train_index, test_index in loo.split(X_array):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X_array[train_index], X_array[test_index]
    y_train, y_test = y[train_index], y[test_index]
    #print(X_train, X_test, y_train, y_test)
    loo_scores.append(rf.score(X_test, y_test))
```

```python
print "Average of leave one out cross validation score:", np.mean(loo_scores)
```

```python
kf = KFold(n_splits=2) 
kf.get_n_splits(X) # returns the number of splitting iterations in the cross-validator
#print(kf) 
KFold(n_splits=2, random_state=None, shuffle=False)
kf_scores=[]

for train_index, test_index in kf.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X_array[train_index], X_array[test_index]
    y_train, y_test = y[train_index], y[test_index]
    #print(X_train, X_test, y_train, y_test)
    kf_scores.append(rf.score(X_test, y_test))
```

```python
print "Average of 2 hold cross validation score:", np.mean(kf_scores)
```
