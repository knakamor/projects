```python
X= pd.read_pickle('articles.pkl')
df=pd.read_pickle('articles.pkl')
#y= X.pop('section_name')
#y=y.unique()
#X.head(2)

vectorizer = TfidfVectorizer(stop_words='english')

vectorized_content = vectorizer.fit_transform(X['content'])

vectorizer.get_feature_names
```

```python
kmeans = KMeans(n_clusters=10, max_iter=100).fit(vectorized_content)
```

```python
kmeans.cluster_centers_
```

```python
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

terms = vectorizer.get_feature_names()
for i in range(10):
    print("Cluster %d:" % i, end='')
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind], end='')
    print()
```

```python
vectorizer2 = TfidfVectorizer(stop_words='english', max_features=100) 
vectorized_content2 = vectorizer2.fit_transform(X['content'])
kmeans2 = KMeans(n_clusters=10, max_iter=100).fit(vectorized_content2)


terms2 = vectorizer2.get_feature_names()
order_centroids2 = kmeans2.cluster_centers_.argsort()[:, ::-1]


for i in range(10):
    print("Cluster %d:" % i, end='')
    for ind in order_centroids2[i, :10]:
        print(' %s' % terms2[ind], end='')
    print()
```

```python
X=vectorizer2.fit_transform(df[:50]['content'])
```

```python
dist=pdist(X.todense())
X3=squareform(dist)
```

```python
z=linkage(X3, 'ward')
```

```python

# calculate full dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel(df[:50]['headline'])
plt.ylabel('distance')
dendrogram(
    z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()


```
